```python
# e.g.
import numpy as np

X = np.linspace(-np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
# 默认配置绘图
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
# 创建一个8*6点（point）的图，并设置分辨率为20
plt.figure(figsize=(8,6), dpi = 80)

# 创建一个新的1*1点子图，接下来的图样绘制在其中的第一块（也是唯一的一块）
plt.subplot(1,1,1)
# 控制余弦曲线，使用蓝色的、连续的、宽度为1（像素）的线条
plt.plot(X, C, color = 'green', linewidth=1.0, linestyle='-')

# 设置横轴的上下限
plt.xlim(-4.0,4.0)
# 设置横轴坐标点
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴的上下限
plt.ylim(-1.0,1.0)
# 设置纵轴坐标点
plt.yticks(np.linspace(-1,1,5,endpoint=True))

plt.show()
```

```python
# 改变线条颜色和粗细
plt.figure(figsize=(10,6), dpi=80)

plt.xlim(-4.0,4.0)
plt.xticks(np.linspace(-4,4,9,endpoint=True))

plt.ylim(-1.0,1.0)
plt.yticks(np.linspace(-1,1,5,endpoint=True))

plt.plot(X, C, color = 'blue', linewidth=2.5, linestyle='-')
plt.plot(X, S, color = 'red', linewidth=2.5, linestyle='-')

plt.show()
```

```python
# 设置图片边界
plt.figure(figsize=(10,6), dpi=80)

# 坐标的设置
xmin, xmax = X.min(),X.max()
ymin, ymax = C.min(),C.max()

dx = (xmax-xmin)*0.2
dy = (ymax-ymin)*0.2

plt.xlim(xmin-dx, xmax+dx)
plt.ylim(ymin-dy, ymax+dy)

plt.plot(X,C,color='blue',linewidth=2.5, linestyle='-')
plt.plot(X,S,color='red',linewidth=2.5, linestyle='-')

plt.show()
```

```python
# 设置轴坐标点的标签

plt.figure(figsize=(10,6), dpi=80)

xmin, xmax = X.min(),X.max()
ymin, ymax = C.min(),C.max()

dx = (xmax-xmin)*0.2
dy = (ymax-ymin)*0.2

plt.xlim(xmin-dx, xmax+dx)
plt.ylim(ymin-dy, ymax+dy)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,+1], [r'$-1$', r'$0$', r'$+1$'])

plt.plot(X,C,color='blue',linewidth=2.5, linestyle='-')
plt.plot(X,S,color='red',linewidth=2.5, linestyle='-')

plt.show()
```

```python
# 移动脊柱（spines）
plt.figure(figsize=(10,6), dpi=80)
# ...

# 添加注释
# 创建一个8×6点的图，dpi设置为80
plt.figure(figsize=(10,6),dpi=80)

# 坐标轴设置
xmin,xmax=X.min(),X.max()
ymin,ymax=C.min(),C.max()

dx=(xmax-xmin)*0.2
dy=(ymax-ymin)*0.2

plt.xlim(xmin-dx,xmax+dx)
plt.ylim(ymin-dy,ymax+dy)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],['$-\pi$','$-\pi/2$','0','$+\pi/2$','$+\pi$'])
plt.yticks([-1,0,1],['$-1$','$0$','$+1$'])

ax=plt.gca()
# 右侧 不显示
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 设置x轴上坐标点在轴下方
ax.xaxis.set_ticks_position('bottom')
# 设置x轴位置
ax.spines['bottom'].set_position(('data',0))
# 设置y轴上坐标点在轴左侧
ax.yaxis.set_ticks_position('left')
# 设置y轴位置
ax.spines['left'].set_position(('data',0))

plt.plot(X,C,color='blue',linewidth=2.5,linestyle='-',label='cosine')
plt.plot(X,S,color='red',linewidth=2.5,linestyle='-',label='sin')

plt.legend(loc='upper left')

t=2*np.pi/3
#余弦的垂线
plt.plot([t,t],[0,np.cos(t)],color='blue',linewidth=2.5,linestyle='--')
# 余弦标注点
plt.scatter([t,],[np.cos(t),],50,color='blue')
# 余弦标注点注视
plt.annotate('$\cos(\frac{2\pi} {3})=-\frac{1}{2}$',xy=(t,np.cos(t)),xycoords='data',xytext=(-90,-50),textcoords='offset points',fontsize=15,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

plt.plot([t,t],[0,np.sin(t)],color='red',linewidth=2.5,linestyle='--')
plt.scatter([t,],[np.sin(t),],50,color='red')
plt.annotate('$\sin(\frac{2\pi}{3})=-\frac{\sqrt{3}}{2}$',xy=(t,np.sin(t)),xycoords='data',xytext=(+10,+30),textcoords='offset points',fontsize=15,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

plt.show()
```
