## Java白皮书关键术语

### 简单性（C++纯净版）

### 面向对象
关注数据（即对象）及对象的借口。（有点像目标导向的思想）

### 网络技能（网络连接）

### 健壮性（robustness)
Java编译器能检测出许多其他语言仅在运行时刻才能检测出的问题\
指针、内存、数据结构之间的关系

### 安全性
e.g.
- 运行时堆栈溢出
- 在自己的处理空间之外破坏内存
- 未经授权读写文件

### 体系结构中立
虚拟机：编译器编译字节码，虚拟机将字节码翻译成机器码。这一中间体系结构使 Java代码可在许多处理器上运行。\
另外，虚拟机还可以检测指令序列的行为，增强其安全性；有些程序还可以快速生成字节码，并动态地增强所运行程序的处理能力

### 可移植性
二进制数据以固定格式进行存储和传输\
作为系统组成部分的类库，定义了可移植的接口（没懂）

### 解释型

### 高性能
即时编译器

### 多线程 —— 带来更好的交互响应和实时行为
Java将多线程的实现交给底层操作系统或线程库来完成

### 动态性
自由添加新方法和实例变量而对客户端没有影响\
当前Java允许程序员知道对象的结构和行为，这对于必须在运行时分析对象的系统来说非常有用

##  Java applet 与 Internet
 applet：在网页中运行Java程序（有虚拟机的安全机制抵御恶意代码的攻击）\
 现动态网页多用 JavaScript 和 falsh实现

## Java的发展简史
Sun公司，消费设备处理能力和内存有限：要求代码紧凑、短小且与平台无关（CPU不同）\
随internet万维网发展（其关键在于将超文本页面转移到屏幕上的浏览器），而当时（90年代）Java已经拥有在客户机/服务器主流模型中所需体系结构，作者表达Gosling（创始人之）认为中立、实时、可靠、安全在工作站环境中并不重要，于是他们决定开发浏览器\
后随数据中心越来越依赖于商业硬件而不是专用服务器，Sun Microsystem于2009年被oracle收购

## 关于Java的常见误解
- Java是一种程序设计语言，HTML是一种描述网页结构的方式。除用于在网页上放置Java applet的html扩展外，二者没有任何共同之处
- Java是一种程序设计语言，XML是一种描述数据的方式。可以使用别的程序设计语言处理XML数据，但Java对XML的处理提供了很好的支持
