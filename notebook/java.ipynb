# 编译项目

## 本地运行
在插件中找到 jetty，双击 jetty:run 以启动项目，然后在浏览器访问：http://localhost:8888 即可看到登录页面。

## 在tomcat部署项目
### [tomcat下载](https://cloud.tencent.com/developer/article/1599252)
### 打包代码
双击 package 执行打包命令，你会在 target 目录下看到生成的 bank-customer-1.0.0-SNAPSHOT.war文件

删除\apache-tomcat-8.5.79\webapps\ROOT目录下的所有文件，然后将打包的war包用解压工具打开，将里面的所有文件复制到ROOT目录下面 \
双击执行 \apache-tomcat-8.5.79\bin\startup.bat 文件，此时浏览器访问：http://localhost:8080 即看到登录页面
